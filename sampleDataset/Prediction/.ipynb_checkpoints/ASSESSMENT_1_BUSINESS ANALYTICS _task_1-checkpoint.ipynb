##### Splitting the data

```python
# Split data into features (X) and target (y)
X = model_df.drop('Attrition', axis=1)
y = model_df['Attrition']
```

```python
# sampling imbalanced data
X_resampled, y_resampled = resample(X, y, random_state=42)
```

```python
# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.36)
```

```python
# Function to plot learning curve for loss, accuracy, and epoch
def plot_loss_accuracy_learning_curve(estimator, X, y, title="Learning Curve"):
    train_sizes, train_losses, test_losses = learning_curve(estimator, X, y, cv=5, scoring="neg_log_loss", n_jobs=-1, train_sizes=np.linspace(0.1, 1.0, 10))
    
    # Convert negative log loss to positive values
    train_losses = -train_losses
    test_losses = -test_losses
    
    # Calculate mean and standard deviation for training and test losses
    train_mean_loss = np.mean(train_losses, axis=1)
    train_std_loss = np.std(train_losses, axis=1)
    test_mean_loss = np.mean(test_losses, axis=1)
    test_std_loss = np.std(test_losses, axis=1)
    
    # Calculate mean accuracy for training and test sets
    train_sizes, train_scores, test_scores = learning_curve(estimator, X, y, cv=5, scoring="accuracy", n_jobs=-1, train_sizes=np.linspace(0.1, 1.0, 10))
    train_mean_accuracy = np.mean(train_scores, axis=1)
    test_mean_accuracy = np.mean(test_scores, axis=1)
    
    # Plot the learning curve for loss and accuracy
    plt.figure(figsize=(10, 6))
    
    # Plot Loss
    plt.subplot(1, 2, 1)
    plt.plot(train_sizes, train_mean_loss, label="Training Loss", color="blue", marker="o")
    plt.fill_between(train_sizes, train_mean_loss - train_std_loss, train_mean_loss + train_std_loss, color="blue", alpha=0.25)
    plt.plot(train_sizes, test_mean_loss, label="Validation Loss", color="red", marker="o")
    plt.fill_between(train_sizes, test_mean_loss - test_std_loss, test_mean_loss + test_std_loss, color="red", alpha=0.25)
    plt.title("Loss Learning Curve")
    plt.xlabel("Training Examples")
    plt.ylabel("Loss")
    plt.legend(loc="best")
    plt.grid()
    
    # Plot Accuracy
    plt.subplot(1, 2, 2)
    plt.plot(train_sizes, train_mean_accuracy, label="Training Accuracy", color="green", marker="o")
    plt.plot(train_sizes, test_mean_accuracy, label="Validation Accuracy", color="purple", marker="o")
    plt.title("Accuracy Learning Curve")
    plt.xlabel("Training Examples")
    plt.ylabel("Accuracy")
    plt.legend(loc="best")
    plt.grid()
    
    plt.suptitle(title)
    plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust layout for better spacing
    plt.show()

```

As there is a target variable with boolean data "Logistic Regression" have been used.

## Logistic Regression

training

```python

LR = LogisticRegression(random_state=42)

LR.fit(X_train,y_train)
y_pred1 = LR.predict(X_train)
```

```python
# Model Accuracy
print("Accuracy:",metrics.accuracy_score(y_train, y_pred1))

print(classification_report(y_train, y_pred1, target_names=["0", "1"]))
cf_matrix= confusion_matrix(y_train, y_pred1)
print(cf_matrix)

y_score = LR.predict_proba(X_train)[:,1]
false_positive_rate, true_positive_rate, threshold = roc_curve(y_train, y_score)
```

validation

```python
y_pred2 = LR.predict(X_test)
```

```python
# Model Accuracy
print("Accuracy:",metrics.accuracy_score(y_test, y_pred2))

print(classification_report(y_test, y_pred2, target_names=["0", "1"]))
cf_matrix= confusion_matrix(y_test, y_pred2)
print(cf_matrix)

y_score = LR.predict_proba(X_test)[:,1]
false_positive_rate, true_positive_rate, threshold = roc_curve(y_test, y_score)
```

```python
plt.plot(false_positive_rate,true_positive_rate)
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
feature_importance = np.abs(LR.coef_[0])

importances = pd.DataFrame(data={
    'Attribute': X_train.columns,
    'Importance':feature_importance
})
importances = importances.sort_values(by='Importance', ascending=False)

importances
```

```python
plt.bar(x=importances['Attribute'], height=importances['Importance'], color='#087E8B')
plt.title('Feature importances obtained from coefficients', size=20)
plt.xticks(rotation='vertical')
plt.show()
```

```python
plot_loss_accuracy_learning_curve(LR, X_train, y_train, title="Logistic Regression Learning Curve")
```
