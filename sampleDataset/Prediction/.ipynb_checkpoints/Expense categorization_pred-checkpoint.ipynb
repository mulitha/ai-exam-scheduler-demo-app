```python
bill_data = df_one_line
```

```python
# Convert labeled_data Item names to uppercase for matching
labeled_data['Item'] = labeled_data['Item'].str.upper()
```

```python
# Create a dictionary mapping items to categories
category_dict = dict(zip(labeled_data['Item'], labeled_data['Category']))
```

```python
# Function to categorize each bill entry
def categorize_bill(items):
    for item in items:  # Check each word in the bill data
        if item in category_dict:
            return category_dict[item]  # Return first match found
    return 'UNKNOWN'  # Default category if no match found

# Apply categorization function
bill_data['Category'] = bill_data['no_punc'].apply(categorize_bill)

# Display result
print(bill_data)
```

## Model Building

```python
# Convert text data (Item names) into numerical features
vectorizer = CountVectorizer(stop_words='english')
```

```python
# Convert lists to strings
bill_data['no_punc'] = bill_data['no_punc'].apply(lambda x: ' '.join(x) if isinstance(x, list) else x)

# Now apply the vectorizer
X_train = vectorizer.fit_transform(bill_data['no_punc'])
```

```python
y_train = bill_data['Category']
```

```python
# Split labeled data for model evaluation
X_train_split, X_test_split, y_train_split, y_test_split = train_test_split(
    X_train, y_train, test_size=0.2, random_state=42
)

# Train the Nave Bayes classifier
model = MultinomialNB()
model.fit(X_train_split, y_train_split)

# Model evaluation
predictions = model.predict(X_test_split)
print("Model Accuracy:", accuracy_score(y_test_split, predictions))
print(classification_report(y_test_split, predictions))
```

```python
import pickle

# Save the model
with open("naive_bayes_model.pkl", "wb") as file:
    pickle.dump(model, file)

# Save the vectorizer
with open("vectorizer.pkl", "wb") as file:
    pickle.dump(vectorizer, file)

print("Model and vectorizer saved successfully!")
```

```python
import re
import pandas as pd
import numpy as np

def extract_receipt_amounts(text):
    # If the input is a list or numpy array, join its elements into a single string.
    if isinstance(text, (list, np.ndarray)):
        text = " ".join(map(str, text))
    
    # Ensure the input is a string.
    if not isinstance(text, str):
        text = str(text)
    
    # Now check if the string is NaN.
    if pd.isna(text):
        return {}
    
    pattern = re.compile(
        r"((?:SUB[\s-]*TOTAL)|(?:NET[\s-]*TOTAL)|TOTAL|AMOUNT)\s+"
        r"(?:LKR|RS)?\s*"
        r"((?:\d+[ ,]*)+\.?\d*)",
        re.IGNORECASE
    )
    
    matches = pattern.findall(text)
    amounts_dict = {}
    
    for raw_label, raw_amount in matches:
        # Normalize the label: remove spaces/hyphens and convert to lowercase.
        label = raw_label.lower().replace(" ", "").replace("-", "")
        
        # Remove commas and spaces from the numeric string.
        clean_amount = raw_amount.replace(",", "").replace(" ", "")
        
        # Attempt to parse as float.
        try:
            # If no decimal point is present and length > 2, assume last 2 digits are decimals.
            if "." not in clean_amount and len(clean_amount) > 2:
                clean_amount = clean_amount[:-2] + "." + clean_amount[-2:]
            parsed_amount = float(clean_amount)
        except ValueError:
            continue
        
        # Append the parsed amount to the dictionary under the normalized label.
        if label not in amounts_dict:
            amounts_dict[label] = []
        amounts_dict[label].append(parsed_amount)
        
    return amounts_dict
```

```python
data_emp_comm_clear["extracted_amounts"] = data_emp_comm_clear["no_punc"].apply(extract_receipt_amounts)
print(data_emp_comm_clear)
```
